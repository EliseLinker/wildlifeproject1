```python
# Simple Charts for CITES data
# Import Dependencies
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# files to used

InputFile = 'CITES_ResultsJRE.csv'
ImporterFile = 'Top10ImpByTrans_JRE.csv'
ExporterFile = 'Top10ExpByTrans_JRE.csv'
ImpQtyFile = 'Top10ImpByQuantity_JRE.csv'
ExpQtyFile = 'Top10ExpByQuantity_JRE.csv'
```

```python
# Simple Pie chart - Importer File (by transaction count)
csvpath = os.path.join(ImporterFile)
importer_df = pd.read_csv(csvpath)
print(importer_df.columns.get_values())
importer_df.head(1)
```

```python
# Import by Trans Count pie chart time!

# Labels for the sections of our pie chart
ImpLabels = importer_df['Import Country']
PieTitle = 'Import Country by Transaction Count'

IBT = importer_df['Transaction Count']
IC = importer_df['Import Country']
ImpTransPie = IBT.plot(kind = 'pie', y = IC, title = PieTitle,
                     autopct="%1.1f%%", explode = None, labels = ImpLabels,
                     colors = None, shadow=False, startangle=210)

plt.show()
```

```python
# and now as a horizontal bar chart

importer_df.plot.barh(stacked=True)

plt.show()
```
